```python
# Install necessary libraries
!pip install requests lxml[html_clean] newspaper3k

# Import libraries
import requests
from newspaper import Article, Config
import time

# GNews API setup
API_KEY = '30af7d389bb77bf04e708e6677e7fba9'  # Replace with your actual GNews API key

# Custom config to mimic a browser
user_config = Config()
user_config.browser_user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'

# ----------- SEARCH ENDPOINT -----------
SEARCH_URL = 'https://gnews.io/api/v4/search'
search_params = {
    'q': 'pizza',       # Search query (required)
    'lang': 'en',                         # Language (default: 'en')
    'country': '',                        # Country code
    'max': 10,                            # Max number of articles (default: 10, max: 100)
    'in': 'title',    # Fields to search in title,description,content
    'from': '',                           # From date (ISO format, e.g. '2024-01-01T00:00:00Z')
    'to': '',                             # To date (ISO format)
    'expanded': 'false',                  # Include publisher info
    'apikey': API_KEY
}

# ----------- TOP HEADLINES ENDPOINT -----------
TOP_HEADLINES_URL = 'https://gnews.io/api/v4/top-headlines'
top_headlines_params = {
    'q': 'pizza',
    'lang': 'en',                         # Language
    'country': 'us',                      # Country
    'topic': '',                          # Topic (e.g. business, science, health, etc.)
    'max': 10,                            # Max articles
    'apikey': API_KEY
}

# Store failed URLs
failed_urls = []

# Call Search Endpoint
print("\n--- SEARCH RESULTS ---")
search_response = requests.get(SEARCH_URL, params=search_params)
if search_response.status_code == 200:
    articles = search_response.json().get('articles', [])
    for i, item in enumerate(articles):
        print(f"{i+1}. {item['title']}")
        print(f"URL: {item['url']}")
        try:
            article = Article(item['url'], config=user_config)
            article.download()
            article.parse()
            print(f"Content:\n{article.text[:2147483647]}...\n")
        except Exception as e:
            print(f"⚠️ Skipping article due to download error: {e}\n")
            failed_urls.append(item['url'])
            continue
else:
    print(f"Search Error: {search_response.status_code} - {search_response.text}")

# Call Top Headlines Endpoint
print("\n--- TOP HEADLINES ---")
top_response = requests.get(TOP_HEADLINES_URL, params=top_headlines_params)
if top_response.status_code == 200:
    headlines = top_response.json().get('articles', [])
    for i, item in enumerate(headlines):
        print(f"{i+1}. {item['title']}")
        print(f"URL: {item['url']}\n")
        try:
            article = Article(item['url'], config=user_config)
            article.download()
            article.parse()
            print(f"Content:\n{article.text[:2147483647]}...\n")
        except Exception as e:
            print(f"⚠️ Skipping article due to download error: {e}\n")
            failed_urls.append(item['url'])
            continue
else:
    print(f"Top Headlines Error: {top_response.status_code} - {top_response.text}")

# Report failed URLs
if failed_urls:
    print("\n--- FAILED URLS ---")
    for url in failed_urls:
        print(url)
```
